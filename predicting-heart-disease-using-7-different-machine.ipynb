#### 1.1 Examining the Project Topic <a id=1></a>

**Understanding Heart Attack (Myocardial Infarction):**

A heart attack, medically termed as "Myocardial Infarction", occurs due to the occlusion of a blood vessel by plaque-like lesions containing cholesterol and fat. These lesions, known as atherosclerotic plaques, are abnormal conditions that develop within the arteries.

When a plaque ruptures or becomes unstable, it can trigger the formation of a blood clot, leading to the obstruction of blood flow to a portion of the heart muscle. Consequently, the affected heart tissue is deprived of oxygen and nutrients.

The blockage of blood flow can result in severe chest pain, known as angina, and if left untreated, it can lead to a heart attack. Without prompt medical intervention, a heart attack can cause irreversible damage to the heart muscle and may even be fatal.

Recognizing the symptoms of a heart attack and seeking immediate medical attention are crucial for improving outcomes and preventing complications.

Here's an explanation of how a heart attack occurs, incorporating the provided text:


**Understanding How a Heart Attack Occurs**

The heart functions as a powerful pump, circulating blood throughout the body approximately 60-80 times per minute while at rest. Despite its role in supplying blood to the entire body, the heart itself requires a steady supply of oxygenated blood to function optimally. This vital supply is facilitated by coronary arteries, which nourish the heart muscle.

However, disruptions in the circulation of these coronary arteries can lead to a condition known as coronary insufficiency. The severity and manifestation of coronary insufficiency can vary based on factors such as the type, degree, and location of arterial narrowing or blockage (stenosis).

Some individuals may experience angina—a type of chest pain—during physical exertion, which subsides with rest. However, in more severe cases, a heart attack (myocardial infarction) can occur. This occurs when there is a sudden and complete occlusion of a coronary artery, depriving a portion of the heart muscle of oxygen-rich blood. The onset of a heart attack is often marked by intense chest pain and may lead to serious complications, including sudden cardiac arrest and death.


**Symptoms of a Heart Attack**

Symptoms of a heart attack can vary but commonly include:

- Severe chest pain or discomfort, often described as pressure, squeezing, or fullness
- Pain or discomfort in other areas of the upper body, such as the arms, back, neck, jaw, or stomach
- Shortness of breath
- Nausea or vomiting
- Cold sweats
- Dizziness or lightheadedness

It's important to note that not everyone experiences the same symptoms, and some individuals, particularly women, may have atypical symptoms. Prompt recognition of symptoms and seeking immediate medical attention are crucial for improving outcomes and preventing further complications.


#### 1.2 Recognizing Variables In Dataset<a id=1.2></a> 

Based on the provided dataset descriptions, here are the definitions of the variables:

1. **age**: Age of the patient in years.
2. **sex**: Sex of the patient. (1 = male; 0 = female)
3. **cp**: Chest pain type.
   - 0: Typical angina (chest pain related decrease blood supply to the heart).
   - 1: Atypical angina (chest pain not related to heart).
   - 2: Non-anginal pain (typically esophageal spasms, non-heart related).
   - 3: Asymptomatic (chest pain not showing signs of disease).
4. **trestbps**: Resting blood pressure in mm Hg on admission to the hospital. (Values above 130-140 are typically cause for concern.)
5. **chol**: Serum cholesterol in mg/dl. (Values above 200 are cause for concern.)
6. **fbs**: Fasting blood sugar level > 120 mg/dl. (1 = true; 0 = false) (Values >126 mg/dL signal diabetes.)
7. **restecg**: Resting electrocardiographic results.
   - 0: Nothing to note.
   - 1: ST-T Wave abnormality, ranging from mild symptoms to severe problems, signaling a non-normal heartbeat.
   - 2: Possible or definite left ventricular hypertrophy, indicating an enlarged heart's main pumping chamber.
8. **thalach**: Maximum heart rate achieved.
9. **exang**: Exercise-induced angina. (1 = yes; 0 = no)
10. **oldpeak**: ST depression induced by exercise relative to rest, assessing the stress of the heart during exercise. An unhealthy heart will experience more stress.
11. **slope**: The slope of the peak exercise ST segment.
    - 0: Upsloping (better heart rate with exercise, uncommon).
    - 1: Flatsloping (minimal change, typical of a healthy heart).
    - 2: Downsloping (signs of an unhealthy heart).
12. **ca**: Number of major vessels (0-3) colored by fluoroscopy. A colored vessel indicates that the doctor can see the blood passing through, with more blood movement indicating better circulation (no clots).
13. **thal**: Thalium stress result.
    - 1, 3: Normal.
    - 6: Fixed defect (previously defective but now okay).
    - 7: Reversible defect (no proper blood movement when exercising).
14. **target**: Presence of heart disease. (1 = yes, 0 = no)

These definitions provide a comprehensive understanding of the variables included in the dataset and their significance for predicting the presence of heart disease.

### 4. Exploratory Data Analysis (EDA)
For our Exploratory Data Analysis (EDA), we'll take it in two main steps:

#### 4.1 Univariate Analysis
We undertake univariate analysis on the dataset's features, based on their datatype:

- **For continuous data:** We employ histograms to gain insight into the distribution of each feature. This allows us to understand the central tendency, spread, and shape of the dataset's distribution.
- **For categorical data**: Bar plots are utilized to visualize the frequency of each category. This provides a clear representation of the prominence of each category within the respective feature.

#### 4.1.1 Numerical Variables Univariate Analysis

```python
# Calculate optimal bin width using Freedman-Diaconis rule
def get_optimal_bin_width(data):
    iqr = np.percentile(data, 75) - np.percentile(data, 25)
    bin_width = 2 * iqr / (len(data) ** (1/3))
    return bin_width

# Set up the subplot
fig, ax = plt.subplots(nrows=2, ncols=3, figsize=(15, 10))

# Loop to plot histograms for each continuous feature
for i, col in enumerate(df[numeric_var].columns):
    x = i // 3
    y = i % 3
    
    # Calculate optimal bin width
    bin_width = get_optimal_bin_width(df[numeric_var][col])
    
    # Plot histogram with KDE
    sns.histplot(data=df[numeric_var], x=col, bins='auto', kde=True, ax=ax[x, y],
                 palette='viridis', edgecolor='none', alpha=0.7, linewidth=1.5)
    
    # Customize axes labels and ticks
    ax[x, y].set_xlabel(col, fontsize=12)
    ax[x, y].set_ylabel('Density', fontsize=12)
    ax[x, y].tick_params(axis='both', which='major', labelsize=10)
    
    # Add mean and median annotations with contrasting colors
    mean_val = df[numeric_var][col].mean()
    median_val = df[numeric_var][col].median()
    ax[x, y].axvline(mean_val, color='orange', linestyle='--', linewidth=1.5)
    ax[x, y].axvline(median_val, color='purple', linestyle='--', linewidth=1.5)
    ax[x, y].text(mean_val, ax[x, y].get_ylim()[1]*0.9, f'Mean: {mean_val:.2f}', fontsize=10, color='orange', ha='center')
    ax[x, y].text(median_val, ax[x, y].get_ylim()[1]*0.85, f'Median: {median_val:.2f}', fontsize=10, color='purple', ha='center')

# Remove empty subplot
ax[1,2].axis('off')

# Set plot title and adjust layout
plt.suptitle('Distribution of Continuous Variables', fontsize=16)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])

# Show plot
plt.show()
```

#### 4.1.2 Categorical Variables Univariate Analysis

```python
# Set up the subplot for a 4x2 layout
fig, ax = plt.subplots(nrows=5, ncols=2, figsize=(15, 18))

# Loop to plot bar charts for each categorical feature in the 4x2 layout
for i, col in enumerate(categoric_var):
    row = i // 2
    col_idx = i % 2
    
    # Calculate frequency percentages
    value_counts = df[col].value_counts(normalize=True).mul(100).sort_values()
    
    # Plot bar chart
    value_counts.plot(kind='barh', ax=ax[row, col_idx], width=0.8, color='blue', alpha=0.5)
    
    # Add frequency percentages to the bars
    for index, value in enumerate(value_counts):
        ax[row, col_idx].text(value, index, f'{value:.1f}%', fontsize=12, weight='bold', va='center', color='black')
    
    # Set x-axis limit and labels
    ax[row, col_idx].set_xlim([0, 100])
    ax[row, col_idx].set_xlabel('Frequency Percentage', fontsize=12)
    
    # Customize title for better readability
    ax[row, col_idx].set_title(f'{col}', fontsize=18, weight='bold')

    # Add grid lines for better visualization
    ax[row, col_idx].grid(axis='x', linestyle='--', alpha=0.7)

# Remove the empty subplot
ax[4, 1].axis('off')

# Set overall title and adjust layout
plt.suptitle('Distribution of Categorical Variables', fontsize=22, weight='bold')
plt.tight_layout(rect=[0, 0.03, 1, 0.95])

# Show plot
plt.show()
```

Their distribution peaks near zero, indicating less ST depression compared to the non-disease category, which has a wider spread. This suggests that ST depression could be a valuable indicator of heart disease risk.

Overall, Maximum Heart Rate Achieved (thalach) appears to have the most significant impact on heart disease status, followed by ST Depression (oldpeak) and Age (age). #### 4.2.2 Categorical Features vs output

```python
# Remove 'output' from the categorical_features
categorical_features = [feature for feature in categoric_var if feature != 'output']
categorical_features
```

```python
from matplotlib.colors import ListedColormap

# Set custom colors
custom_colors = ['#FF7F0E', '#1F77B4']

# Remove 'output' from the categorical_features
categorical_features = [feature for feature in categorical_features if feature != 'output']
fig, ax = plt.subplots(nrows=2, ncols=4, figsize=(15, 10))

for i, col in enumerate(categorical_features):
    
    # Create a cross tabulation showing the proportion of purchased and non-purchased loans for each category of the feature
    cross_tab = pd.crosstab(index=df[col], columns=df['output'])
    
    # Using the normalize=True argument gives us the index-wise proportion of the data
    cross_tab_prop = pd.crosstab(index=df[col], columns=df['output'], normalize='index')

    # Define colormap
    cmp = ListedColormap(custom_colors)
    
    # Plot stacked bar charts
    x, y = i // 4, i % 4
    cross_tab_prop.plot(kind='bar', ax=ax[x, y], stacked=True, width=0.8, colormap=cmp,
                        legend=False, ylabel='Proportion', sharey=True)
    
    # Add the proportions and counts of the individual bars to our plot
    for idx, val in enumerate([*cross_tab.index.values]):
        for (proportion, count, y_location) in zip(cross_tab_prop.loc[val], cross_tab.loc[val], cross_tab_prop.loc[val].cumsum()):
            ax[x, y].text(x=idx-0.3, y=(y_location-proportion)+(proportion/2)-0.03,
                         s=f'{count} ({np.round(proportion * 100, 1)}%)',
                         color="black", fontsize=9, fontweight="bold")
    
    # Add legend
    ax[x, y].legend(title='output', loc=(0.7, 0.9), fontsize=8, ncol=2)
    # Set y limit
    ax[x, y].set_ylim([0, 1.12])
    # Rotate xticks
    ax[x, y].set_xticklabels(ax[x, y].get_xticklabels(), rotation=0)
    
# Set the title for the entire figure
plt.suptitle('Categorical Features vs Output Stacked Barplots', fontsize=22)
plt.tight_layout()
plt.show()
```

**Categorical Features vs Target Stacked Barplots summary:**

- **Chest Pain Type (cp):** Different chest pain types present varied heart disease proportions. Types 1, 2, and 3 have higher heart disease proportions compared to type 0, suggesting chest pain type can influence disease prediction.

- **Exercise Induced Angina (exang):** Absence of exercise-induced angina (0) correlates with higher heart disease presence compared to presence (1), indicating significant impact on the target.

- **Fasting Blood Sugar (fbs):** Distribution between fasting blood sugar > 120 mg/dl (1) and not (0) is similar, suggesting limited impact on heart disease prediction.

- **Slope of Peak Exercise ST Segment (slope):** Slope type 2 has notably higher heart disease presence, indicating its potential significance.
